```python
#import numpy and matplotlib 
import numpy as np, matplotlib.pyplot as plt
%matplotlib inline
```

```python
x = np.arange(0,6)
y = x**2
z = x*2
```

```python
#Create a figure object called fig using plt.figure()
#Use add_axes to add an axis to the figure canvas at [0,0,1,1] as ax
fig = plt.figure()
ax = fig.add_axes([0,0,1,1])
plt.plot(x,y)
ax.set_title('Title goes here')
ax.set_xlabel('X variable')
ax.set_ylabel('Y variable')
```

```python
#Create a figure object and put two axes on it, ax1 and ax2. Located at [0,0,1,1] and [0.2,0.5,.2,.2] respectively
fig1 = plt.figure()
ax1 = fig1.add_axes([0,0,1,1])
ax2 = fig1.add_axes([.2,.5,.2,.2])
ax1.plot(x,y)
ax1.set_title('y = x^2')
ax2.plot(x,z)
ax2.set_title('z = x')
```

```python
#Use subplots to create a plot with 1x2 subplots
plt.subplots(nrows = 1, ncols = 2)
```

```python
#plot (x,y) and (x,z) on the axes with different linewidths and styles
plt.subplots(nrows = 1, ncols = 2)
plt.subplot(1,2,1)
plt.plot(x,y,ls = '--', lw = 5, color = 'red')
plt.subplot(1,2,2)
plt.plot(x,z,ls = 'dashdot',lw = 2, color = 'green')
```
