```python
#import libraries
import numpy as np, pandas as pd, cufflinks as cf
from plotly.offline import download_plotlyjs, init_notebook_mode,plot , iplot
%matplotlib inline
```

```python
#for notebooks
init_notebook_mode(connected = True)
```

```python
#for offline use
cf.go_offline()
```

```python
#create df with fake data
df = pd.DataFrame(np.random.rand(100,4), columns= 'A B C D'.split())
df.head()
```

```python
#use iplot for df
df.iplot(kind = 'scatter', x = 'A', y = 'B',mode = 'markers', size = 10)
```

```python
df2 = pd.DataFrame({'Category':'A B C'.split(), 'Value':[39,98,209]})
df2
```

```python
#create an interactive bar plot for df2
df2.iplot(kind = 'bar', x= 'Category', y = 'Value')
```

```python
#create a boxplot for df
df.iplot(kind = 'box')
```

```python
#create a 3d plot for df3
df3 = pd.DataFrame({'x':[1,2,3,4,5],'y':[10,20,30,20,10],'z':[5,4,3,2,1]})
df3.iplot(kind = 'surface', colorscale = 'rdylbu')
```

```python
#create a spread of columns 'A', 'B' in df
df[['A','B']].iplot(kind = 'spread')
```

```python
#create a histogram of column 'A' in df
df['A'].iplot(kind = 'hist', bins = 50, color = 'blue')
```

```python
#create a bubble plot for 'A' vs 'B' and size = 'C'
df.iplot(kind = 'bubble', x = 'A', y = 'B', size = 'C')
```
