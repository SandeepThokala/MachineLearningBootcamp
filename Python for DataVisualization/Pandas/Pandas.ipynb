```python
#import libraries
import numpy as np, pandas as pd, matplotlib.pyplot as plt
%matplotlib inline
```

```python
#read datasets
df3 = pd.read_csv('df3')
df3.head(3)
```

```python
#plot 'a' vs 'b' using pandas
df3.plot.scatter(x='a',y='b',color='red',figsize=(12,3))
```

```python
#create a histogram on col 'a'
df3['a'].plot.hist()
```

```python
#recreate the same plot with even more bins
df3['a'].plot.hist(bins = 25,color = 'brown')
```

```python
#create a boxplot for columns 'a' and 'b'
df3[['a','b']].plot.box()
```

```python
#create a kde plot for column 'd'
df3['d'].plot.kde(color = 'black',ls = 'dashdot', lw = 3)
```

```python
#Create an area plot of all the columns for just the rows up to 30
df3.iloc[:30].plot.area(alpha = 0.4)
```

```python
#try moving the legend out of the plot
df3.iloc[0:30].plot.area(alpha=0.4)
plt.legend(bbox_to_anchor=(1.02, 0, 0.5, 0.5), loc=6)
```
